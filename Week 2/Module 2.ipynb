# Video 1 - Intro to Supervised ML
## What is a Model ? 
Its a specific mathematical or computational description that expresses the relationship between a set of input variables and one or more outcome variables being studied or predicted.  
In **statistical terms** the input variables are called independent variables and the outcome variables are termed dependent variables.  
In **machine learning** we use term features to refer input or independent variables and target values for output or dependent variables.  
Models can be used to understand and explore the structure of data within given dataset (in unsupervised learning)  
OR can be used to predict the outcome (in supervised learning)

## Supervised Learning 
Learn to predict target values from labelled data  
1. **Classification** : target values are discrete classes  
  * **Binary Classification** : Only two classes - 0 (negative_class) & 1 (postive_class)  
    ex. credit card fraud transaction detection either its fraud or not
  * **Multi-class Classification** : More than 2 output classes and output belongs to only 1 class  
    ex. classifying different fruits
  * **Muti-label Classification** : More than 2 output classes and output belongs to more than 1 class  
    ex. when predicting a given movie category, it may belong to horror, romance, adventure, action, or all simultaneously.
2. **Regression** : target values are continuous values (floating-point/real-valued)  
    ex. predicting price of house
    
Many supervised learning methods have both 'flavors' classification & regression.  
Ex. support vector classifier also has regression version - support vector regression


# Video 2 - Generalization, Underfitting & Overfitting

**Generalization** - Generalization ability refers to an algorithm's ability to give accurate predictions for new, previously unseen data  
**Overfitting** - Models that are too complex for amount of training data (probably less data) available are said to overfit and are not likely to generalize well to new examples  
**Underfitting** - Models that are too simple, that don't even do well on the training data, are said to underfit and also not likely to generalize well. 

## Preamble and Review

````python
%matplotlib notebook
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

np.set_printoptions(precision=2)


fruits = pd.read_table('fruit_data_with_colors.txt')

feature_names_fruits = ['height', 'width', 'mass', 'color_score']
X_fruits = fruits[feature_names_fruits]
y_fruits = fruits['fruit_label']
target_names_fruits = ['apple', 'mandarin', 'orange', 'lemon']

X_fruits_2d = fruits[['height', 'width']]
y_fruits_2d = fruits['fruit_label']

X_train, X_test, y_train, y_test = train_test_split(X_fruits, y_fruits, random_state=0)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
# we must apply the scaling to the test set that we computed for the training set
X_test_scaled = scaler.transform(X_test)

knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train_scaled, y_train)
print('Accuracy of K-NN classifier on training set: {:.2f}'
     .format(knn.score(X_train_scaled, y_train)))
print('Accuracy of K-NN classifier on test set: {:.2f}'
     .format(knn.score(X_test_scaled, y_test)))

example_fruit = [[5.5, 2.2, 10, 0.70]]
example_fruit_scaled = scaler.transform(example_fruit)
print('Predicted fruit type for ', example_fruit, ' is ', 
          target_names_fruits[knn.predict(example_fruit_scaled)[0]-1])
````

# Video 5 - Linear Regression: Least - Squares
## Linear models for regression

A **linear model** is a sum of weighted variables that predicts a target output value given an input data instance  
Ex. predicting housing prices

input instance / feature vector : x = (x<sub>0</sub>, x<sub>1</sub>, .....x<sub>n</sub>)  
predicted output: y = w<sub>0</sub>x<sub>0</sub> + w<sub>1</sub>x<sub>1</sub> + ... + w<sub>n</sub>x<sub>n</sub> + b  
paramters to estimate: 
> * w = (w<sub>0</sub>, w<sub>1</sub>, ....,  w<sub>n</sub>) : feature weights / model coefficients
> * b : constant bias term / intercept 
> * y, w & b : are estimated during regression training process

**Linear Model -**  
* Finds w and b that minimizes the Mean Squared Error(RSS - Residual sum of squares) over the training data between predicted target and actual target values of linear model  
[ Mean Squared Error = (The sum of squared differences between predicted target and actual values) / (No. of training points) ]
* No parameters to control model complexity
* Always represented by straight line

**How are Linear Regression Parameters w, b Estimated?**
* Parameters are estimated from training data.
* There are many different ways to estimate w and b:
> Different methods correspond to different "fit" criteria and goals and ways to control model complexity.

* The learning algorithm finds the parameters that optimize an <u>objective function</u>, typically to minimize some kind of <u>loss function</u> of the predicted target values vs.actual target values.
* One of the loss function is using <u>Least-squares Linear Regression ("Ordinary least-squares"):</u>
> * Find w & b that minimizes the mean squared error of the model (a.k.a RSS): _the sum of squared difference between predicted target and actual target values_
> * The residual sum of squares (RSS) measures the level of variance in the error term, or residuals, of a regression model. The smaller the residual sum of squares, the better your model fits your data; the greater the residual sum of squares, the poorer your model fits your data.
> * No parameters to control model complexity

## Linear regression

# Video 6 - Linear Regression: Ridge, Lasso, and Polynomial regression
## Ridge regression

* Ridge regression learns w, b using the same least-squares criterion but adds a penalty for large variations in w parameters (as shown in red box below).



* This penalty is called <u>regularization</u>. Regularization prevents overfitting by restricting the model, typically to reduce complexity. Ridge regression uses <u>L2 regularization</u>: minimize sum of squares of w entries.

* The influence of the regularization term is controlled by the non negative $\alpha$ parameter (a hyperparameter) [default 1.0].
* Higher alpha means more regularization and simpler models
* By tuning the value of the hyperparameter $\alpha$, you can control the trade-off between fitting the data well (minimizing RSS) and keeping the coefficients small (minimizing the regularization term). A larger $\alpha$ results in more regularization, which may lead to simpler models with smaller coefficients. 
* The choice of $\alpha$ is typically determined through techniques like cross-validation.
* The goal of Ridge Regression is to find the values of the regression coefficients (w & b) that minimize the cost function. The L2 regularization term (shown in red box) encourages the coefficients to be small, which helps prevent overfitting and reduces the sensitivity of the model to multicollinearity.
* The solution for Ridge Regression can be found through various optimization techniques, such as gradient descent or closed-form solutions. The final coefficients w & b are often referred to as "shrunken" or "regularized" coefficients because they are penalized to be smaller than in ordinary least squares regression.
* The practical effect of using ridge regression is to find the future weights, wi, that fit the data well in a least square sense, and that set lots of the feature weights to values that are very small. We don't see this effect with a single variable linear regression example, but for regression problems with dozens or hundreds of features, the accuracy improvement from using regularized linear regression like ridge regression can be significant.

````python
from sklearn.linear_model import Ridge
X_train, X_test, y_train, y_test = train_test_split(X_crime, y_crime,
                                                   random_state = 0)

linridge = Ridge(alpha=20.0).fit(X_train, y_train)

print('Crime dataset')
print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
````

## The Need for Feature Normalization  
* Important for some ML methods that all features are on same scale (e.g. faster convergence in learning, more uniform or 'fair' influence for all weights)
> * e.g. reguarized regression, k-NN, support vector machines, neural networks, etc
* Can also depend on data.
* For now we are doing MinMax Scaling. Using MinMax Scaling the data set is rescaled such that all feature values are in the range [0, 1]
> * For each feature $x_i$: compute the min value $x_i^{MIN}$ and the max value $x_i^{MAX}$ achieved across all instances in training set.
> * For each feature: transform a given feature $x_i$ value to a scaled version $x_i^{scaled}$ using the formula - 
$$x_i^{scaled} =  (x_i - x_i^{MIN}) / (x_i^{MAX} - x_i^{MIN})$$

### SciKit Learn MinMaxScaler
* MinMaxScaler rescales the data set such that all feature values are in the range [0, 1]  

### Feature Normalization
* The test set must use identical scaling to the training set
* Fit the scaler using the training set then apply the same scaler to transform the test set.
* Do not scale the training and test sets using different scalers: this could lead to random skew in the data.
* Do not fit the scaler using any part of the test data: referencing the test data can lead to a form of data leakage.

## Ridge regression with feature normalization

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

from sklearn.linear_model import Ridge
X_train, X_test, y_train, y_test = train_test_split(X_crime, y_crime,
                                                   random_state = 0)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

linridge = Ridge(alpha=20.0).fit(X_train_scaled, y_train)

print('Crime dataset')
print('ridge regression linear model intercept: {}'
     .format(linridge.intercept_))
print('ridge regression linear model coeff:\n{}'
     .format(linridge.coef_))
print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train_scaled, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test_scaled, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
````

### Ridge regression with regularization parameter: alpha
* In general regularization works especially well when you have relatively small amounts of training data compared to the number of features in your model.
* Regularization becomes less important as amount of training data increases 

````python
print('Ridge regression: effect of alpha regularization parameter\n')
for this_alpha in [0, 1, 10, 20, 50, 100, 1000]:
    linridge = Ridge(alpha = this_alpha).fit(X_train_scaled, y_train)
    r2_train = linridge.score(X_train_scaled, y_train)
    r2_test = linridge.score(X_test_scaled, y_test)
    num_coeff_bigger = np.sum(abs(linridge.coef_) > 1.0)
    print('Alpha = {:.2f}\nnum abs(coeff) > 1.0: {}, \
r-squared training: {:.2f}, r-squared test: {:.2f}\n'
         .format(this_alpha, num_coeff_bigger, r2_train, r2_test))
````

## Lasso regression
* Lasso regression is another form of regularized linear regression that uses an L1 regularization penalty for training (instead of ridge's L2 penalty) (as shown in red box below)  
> 
* L1 penalty : Minimize the sum of absolute values of the coefficients
* This has the effect of setting parameter weights in w to <u>zero</u> for the least influential variables. This is called a <u>sparse</u> solution: a kind of feature selection
* The hyperparameter $\alpha$ controls amount of L1 regularization (default 1.0). It is a non-negative value that controls the amount of regularization applied to the model. Higher values of Î± lead to stronger regularization.
* By tuning the value of the hyperparameter $\alpha$, you can control the trade-off between fitting the data well (minimizing RSS) and achieving sparsity in the model. A larger $\alpha$ results in more regularization and a sparser model. The choice of $\alpha$ is often determined through techniques like cross-validation.
* The goal of Lasso Regression is to find the values of the regression coefficients w & b that minimize the cost function. The L1 regularization term (shown in red box) encourages the coefficients to be small and, importantly, encourages some of them to be exactly zero. This property makes Lasso Regression useful for feature selection, as it can effectively eliminate less important features by setting their corresponding coefficients to zero.
* The solution for Lasso Regression can be found through various optimization techniques, such as coordinate descent or subgradient methods. The final coefficients w & b obtained through Lasso Regression will have a sparse nature, meaning that many coefficients may be exactly zero.

````python
from sklearn.linear_model import Lasso
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

X_train, X_test, y_train, y_test = train_test_split(X_crime, y_crime,
                                                   random_state = 0)

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

linlasso = Lasso(alpha=2.0, max_iter = 10000).fit(X_train_scaled, y_train)
'''
A lasso regression has a unique optimum, but the solver is a sort of gradient descent algorithm, 
so you'll never actually reach the minimum. tol controls how close you want to be: the smaller tol, 
the more accurate your final solution will be, but the longer it will take. 
max_iter controls how many steps you'll take in the gradient descent before giving up. 
The algorithm will stop when either updates are within tol or you've run for max_iter many steps; 
if the latter, you'll get a warning saying that the model hasn't converged (to within tol).

So, set tol to your liking, and set max_iter according to your computational resources. 
Usually, go with the defaults and increase max_iter (and/or change solver, or scale your data if you haven't yet) 
if you get convergence warnings.
'''

print('Crime dataset')
print('lasso regression linear model intercept: {}'
     .format(linlasso.intercept_))
print('lasso regression linear model coeff:\n{}'
     .format(linlasso.coef_))
print('Non-zero features: {}'
     .format(np.sum(linlasso.coef_ != 0)))
print('R-squared score (training): {:.3f}'
     .format(linlasso.score(X_train_scaled, y_train)))
print('R-squared score (test): {:.3f}\n'
     .format(linlasso.score(X_test_scaled, y_test)))
print('Features with non-zero weight (sorted by absolute magnitude):')

for e in sorted (list(zip(list(X_crime), linlasso.coef_)),
                key = lambda e: -abs(e[1])):
    if e[1] != 0:
        print('\t{}, {:.3f}'.format(e[0], e[1]))
````

### Lasso regression with regularization parameter: alpha

````python
print('Lasso regression: effect of alpha regularization\n\
parameter on number of features kept in final model\n')

for alpha in [0.5, 1, 2, 3, 5, 10, 20, 50]:
    linlasso = Lasso(alpha, max_iter = 10000).fit(X_train_scaled, y_train)
    r2_train = linlasso.score(X_train_scaled, y_train)
    r2_test = linlasso.score(X_test_scaled, y_test)
    
    print('Alpha = {:.2f}\nFeatures kept: {}, r-squared training: {:.2f}, \
r-squared test: {:.2f}\n'
         .format(alpha, np.sum(linlasso.coef_ != 0), r2_train, r2_test))
````

## When to use ridge vs lasso regression :

* **Use Ridge Regression When:**

> * Many features with small/medium sized effects
> * <u>Multicollinearity is a Concern:</u> Ridge Regression is particularly useful when you have multicollinearity in your dataset, meaning that independent variables are highly correlated with each other. It helps by shrinking the coefficients of correlated variables towards each other.
> * <u>You Want to Retain All Features:</u> Ridge Regression tends to keep all the features in the model, although it may reduce the coefficients close to zero without setting them exactly to zero. If you believe that all features are relevant to your prediction task, Ridge can be a good choice.
> * <u>You Prefer a Smooth Shrinkage:</u> Ridge Regression applies a smooth shrinkage to the coefficients, which means that all coefficients are reduced proportionally. This can be beneficial when you have a large number of variables and you want to reduce their impact on the model without completely eliminating any.

* **Use Lasso Regression When:**

> * Only a few features with medium/large effect
> * <u>Feature Selection is Important:</u> Lasso Regression is a feature selection technique as well as a regularization method. It can set the coefficients of less important features to exactly zero, effectively removing them from the model. If you have a large number of features and suspect that only a subset of them is relevant, Lasso can help you automatically select the most important ones.
> * <u>You Want a Sparse Model:</u> Lasso tends to produce sparse models with only a subset of the features having non-zero coefficients. This can lead to simpler and more interpretable models, which is valuable in situations where you want to understand the most influential variables.
> * <u>There is a Budget Constraint:</u> In some cases, you may have a budget constraint that limits the number of variables you can include in your model. Lasso can help you automatically select the most important variables while keeping the model within budget.

* In practice, the choice between Ridge and Lasso often depends on the nature of your data and your specific goals. You can also consider **Elastic Net Regression**, which combines both Ridge and Lasso penalties, offering a balance between feature selection and smooth coefficient shrinkage. Cross-validation is a valuable tool for selecting the appropriate regularization strength (Î± in Ridge and Lasso) and determining which method performs better on your dataset.

## Polynomial regression  



* Generate new features consisting of all polynomial combinations of the orignal two features ($x_0$, $x_1$).
* The degree of polynomial specifies how many variables participate at a time in each new feature (above example : degree 2)
* This is still a weighted linear combination of features, so it's <u>still a linear model</u>, and can use same least-squares estimation method for $w$ and $b$

### Polynomial Features with Linear Regression 

* Why would we want to transform our data this way?
> * To capture interactions between the original features by adding them as features to the linear model.
> * To make a classification problem easier (we'll see this later).
* More generally, we can apply other non-linear transformations to create new features
> * (Technically, these are called non-linear basis functions)
* Beware of polynomial feature expansion with high as this can lead to complex models that overfit
> * Thus, polynomial feature expansion is often combined with a regularized learning method like ridge regression.

## Linear Support Vector Machines
* Linear Support Vector Machines (SVMs) are a type of supervised machine learning algorithm used for classification tasks. They are particularly powerful when the decision boundary between classes is linear or close to linear.
* SVMs are primarily used for **binary classification**, where the goal is to classify data points into one of two classes (e.g., yes/no, spam/ham, positive/negative). However, they can be extended to multiclass classification using techniques like one-vs-one or one-vs-all.
* Linear SVMs work by finding a hyperplane that best separates the data points of one class from those of the other class. This hyperplane is the decision boundary, and it aims to maximize the margin between the two classes. The **margin** is the distance between the hyperplane and the nearest data points from each class.
* **Support vectors** are the data points that are closest to the decision boundary. These are the critical points that influence the position and orientation of the decision boundary. SVMs are named after these support vectors because they **"support"** the definition of the hyperplane.
* The primary objective of a Linear SVM is to maximize the margin while minimizing the classification error. The margin is maximized by finding the hyperplane that has the maximum distance from the support vectors of both classes. This margin-maximizing hyperplane is also called the **"maximum-margin hyperplane."**
* In real-world datasets, it's common to have some overlapping data points or outliers that cannot be perfectly separated by a linear hyperplane. In such cases, Linear SVMs use a "soft margin" approach, which allows for some misclassification to achieve better generalization. The trade-off between maximizing the margin and allowing some misclassification is controlled by a hyperparameter called the **"C" parameter**. Its <u>L2 Regularization</u> parameter. It is positive value. The strength of the regularization is inversely proportional to C
* While Linear SVMs work well for linearly separable data, they can be extended to handle nonlinearly separable data by using a technique called the "kernel trick" (**kernel parameter**) Kernels allow SVMs to implicitly transform the data into a higher-dimensional space where it becomes linearly separable. Common kernels include the polynomial kernel and the radial basis function (RBF) kernel.

# Video 9 - Multi-class classification with linear models
In multi-class classification with scikit-learn, there are two primary strategies for training binary classifiers for each class:

1. **One-vs-Rest (OvR) or One-vs-All (OvA)**:
   - In this approach, a separate binary classifier is trained for each class.
   - For each classifier, one class is treated as the positive class, and all the other classes are treated as the negative class.
   - During prediction, each binary classifier makes a decision, and the class associated with the classifier that produces the highest confidence or probability is considered the predicted class for the input data point.
   - This strategy is commonly used for multi-class problems and is the default in scikit-learn for most classifiers.

2. **One-vs-One (OvO)**:
   - In this approach, a binary classifier is trained for each pair of classes (i.e., one classifier for class 0 vs. class 1, another for class 0 vs. class 2, and so on).
   - During prediction, each binary classifier produces a vote for one of the two classes it was trained on.
   - The class that receives the most votes across all binary classifiers is considered the predicted class for the input data point.
   - While this strategy can be used for multi-class classification, it requires more binary classifiers than the OvR approach (N * (N-1) / 2 classifiers for N classes), making it less efficient in terms of training time and memory usage.

Scikit-learn provides a unified interface for both of these strategies, allowing you to choose the one that best suits your problem. The choice between OvR and OvO often depends on the specific algorithm, the size of the dataset, and the computational resources available. In most cases, OvR is a good default choice and works well for many multi-class classification tasks.

# Video 10 - Kernelized Support Vector Machines

In Support Vector Machine (SVM) algorithms, the term "kernel" refers to a mathematical function that is used to transform the input data into a higher-dimensional space. Kernels play a crucial role in SVMs, particularly in the context of Support Vector Classification (SVC) and Support Vector Regression (SVR). The kernel trick allows SVMs to find a decision boundary (or regression function) in this higher-dimensional space, which can be more expressive and may lead to better separability of data points.

In the context of SVC (Support Vector Classification), the kernel is used to map the original input data from a lower-dimensional space to a higher-dimensional space, where it might be easier to find a hyperplane (decision boundary) that can separate the different classes. The primary idea is that by applying a suitable kernel function, data points that are not linearly separable in the original space can become linearly separable in the higher-dimensional space.

Commonly used kernel functions in scikit-learn's SVC include:

* **Linear Kernel:** It represents a linear transformation and is suitable for <u>linearly</u> separable data.

* **Polynomial Kernel:** The polynomial kernel maps data into a <u>higher-dimensional</u> space using a <u>polynomial function</u>. It can capture more complex relationships between data points.

* **RBF (Radial Basis Function) Kernel:** The RBF kernel, also known as the <u>Gaussian kernel</u>, uses a Gaussian-like function to map data into an <u>infinite-dimensional</u> space. It is useful for capturing non-linear relationships and is a popular choice when the decision boundary is not expected to be linear. <u>Default Kernel in SVC Module.</u>

* **Sigmoid Kernel:** The sigmoid kernel uses a <u>sigmoid function</u> to map data into a <u>higher-dimensional</u> space. It can be useful in some scenarios, but it's less commonly used compared to linear, polynomial, and RBF kernels.

## Support Vector Machine with RBF kernel: gamma parameter
In the context of Support Vector Classification (SVC) in scikit-learn and many other implementations of Support Vector Machines (SVMs), the gamma parameter is a hyperparameter that controls the shape of the decision boundary or the influence of individual training samples on the decision boundary. It specifically affects the behavior of the Radial Basis Function (RBF) kernel, one of the most commonly used kernels in SVMs.

Here's how the gamma parameter works:

* **Gamma and RBF Kernel:** When you use the RBF kernel in an SVM, the gamma parameter defines the shape of the decision boundary. A smaller gamma value makes the decision boundary more flexible and may result in a smoother boundary, while a larger gamma value makes the boundary more rigid and may lead to a more complex and tightly fit decision boundary.

* **Influence of Training Samples:** The gamma parameter also affects how much each training sample influences the decision boundary. A smaller gamma value gives more weight to distant training samples, which means that the decision boundary is influenced by a larger region of the training data. In contrast, a larger gamma value gives more weight to nearby training samples, resulting in a decision boundary that is influenced primarily by the closest data points.

* **Overfitting vs. Underfitting:** Tuning the gamma parameter is crucial for controlling the trade-off between overfitting and underfitting in an SVM. <u>A smaller gamma value can lead to underfitting because the model may be too smooth and fail to capture the fine details in the data. On the other hand, a larger gamma value can lead to overfitting because the model may become too sensitive to noise in the training data.</u>

* **Grid Search and Cross-Validation:** To find an appropriate value for the gamma parameter, it's common to perform a grid search over a range of possible values and use cross-validation to evaluate the model's performance with different gamma settings. The GridSearchCV function in scikit-learn is often used for this purpose.

# Video 11 - Cross-validation

Cross-validation is a widely used technique in machine learning for assessing the performance and generalization ability of a predictive model. It helps to estimate how well a model will perform on unseen data. The primary goal of cross-validation is to provide a more robust evaluation of a model's performance than a single train-test split.

<u>Note</u> - 
* Uses multiple train-test splits, not just a single one
* Each split used to train & evaluate a separate model

**Why is this better?**
* The accuracy score of a supervised learning method can vary, depending on which samples happen to end up in the training set.
* Using multiple train-test splits gives more stable and reliable estimates for how the classifier is likely to perform on average.
* Results are averaged over multiple different training sets instead of relying on a single model trained on a particular training set.

**How cross-validation is used in practice ?**
* Cross-validation gives a more stable performance estimate of a given specific model so that it can be reliably compared with a different model.
> * Compute the mean evaluation metric across CV folds: an estimate, not a guarantee!
> * e.g. are SVMs better than Naive Bayes on this data, for this metric ?
> * Then train a final production classifier with the best model using ALL the data.
* Typically cross-validation is NOT used to PRODUCE a model e.g. by a averaging the resulting models (via the parameters) from each fold
> * This is called <u>model averaging</u> and can be done in certain special cases, like with certain methods if the loss function is convex
* BLUF: for reliable comparison of an evaluation metric across two different model types, use k-fold cross-validation, NOT a single train-test split.

**Cross-validation for model evaluation vs model tuning**
* <u>Evaluate and compare</u> the accuracy/score of different models with k-fold cross-validation (train/test splits)
* <u>Tuning a single model</u> (e.g. find the optimal hyperparameters) uses a slightly different setup: train/<u>validate</u>/test splits

## Leave-one-out cross-validation (LOOCV)
* **Leave-one-out cross-validation (LOOCV)** is a configuration of k-fold cross-validation where k is set to the number of examples in the dataset. In LOOCV, the function approximator is trained on all the data except for one point and a prediction is made for that point. This is repeated N separate times.
* The LOOCV procedure is appropriate when you have a small dataset or when an accurate estimate of model performance is more important than the computational cost of the method.

## A note on performing cross-validation for more advanced scenarios.

* In some cases (e.g. when feature values have very different ranges), we've seen the need to scale or normalize the training and test sets before use with a classifier. The proper way to do cross-validation when you need to scale the data is *not* to scale the entire dataset with a single transform, since this will indirectly leak information into the training data about the whole dataset, including the test data, leading to **data leakage**.

# Video 12 - Decision Trees

A **Decision Tree** is a widely used machine learning algorithm for both classification and regression tasks. It's a supervised learning method that creates a hierarchical structure that breaks down a data set into smaller and smaller subsets based on certain conditions or features, eventually leading to a prediction or decision.

- **Nodes**:
   - <u>Root Node</u>: Represents the entire dataset.
   - <u>Internal Nodes</u>: Contain conditions for splitting data.
   - <u>Leaf Nodes (Terminal Nodes)</u>: Provide the final prediction or decision.

- **Edges (Branches)**: Connect nodes, guiding the decision process.

- **Splitting Criteria**: At internal nodes, a feature-based condition is chosen to divide the data into subsets.

- **Splitting Process**: The data is recursively split into subsets, creating child nodes based on the chosen criteria.

- **Pruning**: To avoid overfitting, nodes are pruned if they don't significantly improve predictive performance.

- **Classification**: Leaf nodes represent class labels, determining the outcome for an instance by traversing the tree.

- **Regression**: Leaf nodes represent numerical predictions.

- **Splitting Criteria**: Common criteria include Gini impurity (classification), entropy (classification), and mean squared error (regression).

- **Feature Importance**: Decision trees offer insights into feature importance, highlighting the most influential attributes.

## Advantages & Limitations
**Advantages**:
- Easy to understand and interpret.
- Handles categorical and numerical data.
- Non-parametric, making no assumptions about data distribution.
- Captures non-linear relationships.
- No feature normalization or scaling typically needed.
- Work well with datasets using a mixture of feature types (continuous, categorical, binary)

**Limitations**:
- Prone to overfitting, especially with deep and complex trees even after tuning.
- Sensitive to minor data variations.
- Bias toward dominant classes.
- Limited expressiveness for complex relationships.
- Usually need an ensemble of trees for better generalization performance.

Ensemble methods, like Random Forests and Gradient Boosting, are often used to address decision trees' limitations, improving robustness and accuracy.

## Are Decision Trees Always Binary?

Decision trees are not always binary; they can be either binary or multiway (also called multi-ary or multi-class) trees, depending on how they are constructed and the nature of the problem they are intended to solve.

Here's an explanation of both types:

1. **Binary Decision Trees**:
   - In a binary decision tree, each internal node has exactly two child nodes.
   - The splitting criteria at each internal node results in two branches, typically based on a binary decision (e.g., true or false).
   - Binary decision trees are most commonly used in classification and regression tasks.

2. **Multiway (Multi-ary) Decision Trees**:
   - In multiway decision trees, each internal node can have more than two child nodes.
   - The splitting criteria at internal nodes can result in multiple branches, corresponding to multiple categories or outcomes.
   - Multiway decision trees are particularly useful for problems with more than two classes or categories.

In practice, the choice between binary and multiway decision trees depends on the specific problem you are trying to solve. For binary classification problems (e.g., yes/no, spam/ham), binary decision trees are often sufficient. However, when dealing with multi-class classification problems (e.g., classifying images into various objects or identifying the type of disease in medical diagnoses), multiway decision trees may be more appropriate.

It's important to note that while binary trees are simpler and easier to interpret, they might require more levels of depth to handle multi-class problems effectively, potentially leading to overfitting. Multiway trees can directly represent multi-class problems, potentially leading to more efficient models. The choice depends on the problem's complexity, the data, and the desired interpretability of the model.   

## Classification vs. Regression Decision Trees

Decision trees can be used for both classification and regression tasks, and they have distinct characteristics for each:

1. **Classification Decision Trees**:

   - Used for tasks where the output is a categorical label or class (e.g., spam/ham, dog/cat).
   - The leaf nodes of a classification tree represent class labels.
   - Splitting criteria are based on impurity measures like <u>Gini impurity or entropy</u>.
   - At each internal node, the tree makes decisions that separate data into different classes.
   - The final prediction is the <u>class label</u> of the leaf node reached.

2. **Regression Decision Trees**:

   - Used for tasks where the output is a numerical value (e.g., predicting house prices, temperature).
   - The leaf nodes of a regression tree represent numerical predictions.
   - Splitting criteria are based on <u>reducing the mean squared error (MSE)</u>.
   - At each internal node, the tree makes decisions that divide data based on numerical features.
   - The final prediction is the <u>average of the numerical values</u> in the leaf node.

Decision trees can be adapted for either classification or regression by modifying the splitting criteria and the type of prediction made at the leaf nodes.

## Controlling the Model Complexity of Decision Trees

Controlling the complexity of decision trees is crucial to prevent overfitting and to build models that generalize well. There are several strategies to manage the complexity of decision trees, and scikit-learn provides relevant parameters to control these aspects:

1. **Maximum Depth (Max Depth):**
   - Limit the depth of the decision tree to a specific number of levels ([`max_depth` parameter in scikit-learn]).
   - Can prevent overfitting but may lead to underfitting if set too low.  


2. **Minimum Samples per Leaf:**
   - Set a minimum number of samples allowed in a leaf node ([`min_samples_leaf` parameter in scikit-learn]).
   - Useful to prevent nodes with very few samples and to avoid further splitting, which could lead to overfitting.  


3. **Minimum Samples per Split:**
   - Specify a minimum number of samples required to perform a split ([`min_samples_split` parameter in scikit-learn]).
   - Prevents splits that would result in small child nodes.  


4. **Maximum Number of Leaf Nodes:**
   - Restrict the maximum number of leaf nodes in the tree ([`max_leaf_nodes` parameter in scikit-learn]).
   - Helps control tree complexity.  


5. **Maximum Features:**
   - Limit the number of features considered when deciding on the best split ([`max_features` parameter in scikit-learn]).
   - Useful for reducing model complexity when working with high-dimensional data.  


6. **Pruning:**
   - Scikit-learn uses Cost-Complexity Pruning, which is based on the Minimal Cost-Complexity Pruning algorithm.
   - It automatically prunes the tree based on the `ccp_alpha` parameter.
   - Post-pruning or pruning the tree after it's fully grown can help simplify complex trees.
   - Pruning methods remove nodes that do not significantly improve the model's performance.
   - One strategy to prevent overfitting is to prevent the tree from becoming really detailed and complex by stopping its growth early. This is called pre-pruning. Another strategy is to build a complete tree with pure leaves but then to prune back the tree into a simpler form. This is called post-pruning or sometimes just pruning. The decision tree implementation and scikit-learn only implements pre-pruning.  

Balancing these parameters is essential for creating decision trees that are neither too complex (overfit) nor too simple (underfit). Properly tuned decision trees can provide a good balance between bias and variance, resulting in accurate and generalizable models. 

In practice, adjusting only one of (`max_depth`, `min_samples_leaf` and `max_leaf_nodes`) is enough to reduce overfitting. In effect, setting this parameter will indirectly influence the other two parameters and vice versa. So in practice, adjusting only one of these trees is typically enough to control most overfitting. Although even with the most optimized parameter settings, individual decision trees will still tend to overfit.

## Feature importance

**Feature Importance in Decision Trees**

Understanding feature importance is crucial when working with decision trees. Feature importance indicates the contribution of each feature to the model's decisions. In decision trees, feature importance is typically determined based on the following principles:

- Features used near the root of the tree are generally more important as they have a significant impact on the initial splits.
- Features used in multiple splits are often more important, indicating their relevance in the decision process.
- Features that lead to pure leaf nodes (all samples in the leaf belong to the same class in classification or have similar values in regression) tend to be highly important.
- A number between 0 and 1 assigned to each feature :
> - Feature importance of 0 -> the feature was not used in prediction.
> - Feature importance of 1 -> the feature predicts the target perfectly.
> - All feature importancesare normalized to sum to 1.

Scikit-learn provides a feature importance score for each feature in decision trees. You can access these scores using the `feature_importances_` attribute after fitting the decision tree model.

It's important to note that feature importance is a relative measure within the context of the specific decision tree model. High feature importance does not necessarily imply causality but indicates the features that had the most influence on the model's decisions.

# Video 13 - One Hot Encoding [OHE]

One-hot encoding is a technique used in data preprocessing to convert categorical variables into a numerical format that machine learning algorithms can work with. Categorical variables are those that represent categories, such as "red," "green," "blue," or "dog," "cat," "bird."

One-hot encoding involves creating binary columns (0 or 1) for each category in the categorical variable. Each category becomes a new column, and a 1 is placed in the column corresponding to the category of a particular data point, while all other columns get a 0. 

**Many predictor cannot take categorical input directly:**
- Ordinary Least Squares Linear Regression
- Logistic Regression
- Support Vector Machines
  
**But some can:**
- Decision trees and tree-based predictors

<u>Note</u>: In statistics, called a 'dummy' encoding, '1 of k' encoding, or indicator variable.

## Here's how one-hot encoding works:

1. Identify the categorical variable you want to encode.
2. For each unique category within that variable, create a new binary column.
3. Assign a 1 in the appropriate column for each data point that corresponds to that category; assign 0 to the other columns.

One-hot encoding is essential for categorical variables when using machine learning models. It prevents the model from interpreting ordinal or nominal categories as having any ordinal relationship. Scikit-learn provides tools like `OneHotEncoder` to perform one-hot encoding efficiently.

## Dense vs. Sparse Matrix

| **Aspect**                        | **Dense Matrix**                                        | **Sparse Matrix**                                          |
| :------------------------------ | :--------------------------------------------------- | :------------------------------------------------------- |
| Sparsity                      | Contains mostly non-zero values                   | Contains mostly zero values                             |
| Data Storage                  | Stores all elements explicitly                      | Stores only non-zero elements and their positions        |
| Memory Usage                  | Higher memory usage                                 | Lower memory usage                                      |
| Computational Efficiency      | Faster arithmetic operations                        | Slower arithmetic operations due to handling zero values |
| Typical Use Cases             | Commonly used for small to moderately sized datasets with low sparsity | Commonly used for large datasets with high sparsity (e.g., in natural language processing and scientific computing) |
| Example                      | Dense_Matrix =<br>[[1, 0, 2],<br>[0, 3, 0],<br>[4, 0, 5]]                   | Sparse_Matrix =<br>[[1, 0, 2],<br>[0, 0, 0],<br>[4, 0, 5]]                             |

**Sparsity(S)** of matrix is calculated using the formula: 
> `S = Z / T`  
> where Z = number of zero elements &   
> T =  elements in the matrix (T)

If the sparsity (S) is close to 1, the matrix is considered <u>sparse</u>. If it's close to 0, the matrix is considered <u>dense</u>. 

## OneHotEncoder vs. LabelBinarizer in scikit-learn

Both `OneHotEncoder` and `LabelBinarizer` in scikit-learn are used for converting categorical variables into a numerical format. However, they have distinct purposes and behaviors:

### `sklearn.preprocessing.OneHotEncoder`

`sklearn.preprocessing.OneHotEncoder` is a class provided by scikit-learn (sklearn) for converting categorical features into one-hot encoded representations. It is a fundamental tool in data preprocessing for machine learning, especially when working with categorical variables.  It follows the one-hot encoding technique, which creates binary columns (containing 0 or 1) for each category. Each category is mutually exclusive, and only one of the cell of the column in the binary representation is 1 (hot) for each data point.

Key aspects of `OneHotEncoder`:

- **Input**: The class takes categorical features as input, and these features are typically stored as integers or strings. Each feature can represent a category with multiple levels (e.g., "red," "green," "blue").

- **Output**: It produces a binary matrix where each row corresponds to a data point, and each column represents a category from the input feature. A 1 in a column indicates that the data point belongs to that category, while 0 indicates that it does not. Produces **sparse** matrix by deault

- **Usage**: 
> - Commonly used in data preprocessing to convert categorical features into a format suitable for machine learning models that expect numerical input.
> - It is used to encode **features** (`X`)

- **Methods**:
> - `fit`: finds unique values per feature column.
> - `transform`: generates the 1-hot encoding.

Here's a basic example of how to use `OneHotEncoder` in scikit-learn:


````python
from sklearn.preprocessing import OneHotEncoder

# Sample categorical feature
colors = ['red', 'green', 'blue', 'red']

# Create an instance of OneHotEncoder
encoder = OneHotEncoder(sparse=False)

print(np.array(colors).reshape(-1, 1))

# Fit and transform the feature
encoded_colors = encoder.fit_transform(np.array(colors).reshape(-1, 1))

encoded_colors
````

- **Methods**:

2. Fit the OneHotEncoder using the training data
3. Use transform to apply this same OneHotEncoder to both training and test data. This ensures that:
> a. Dummy variables match across train and test data  
> b. All categories are represented.

* Options to inculde -
