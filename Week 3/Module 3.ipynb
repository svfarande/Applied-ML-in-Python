# Video 1 - Model Evaluation & Selection

Model evaluation and selection are crucial steps in the machine learning workflow. They involve assessing the performance of different models and choosing the one that best fits the problem at hand. This process is essential for building robust and accurate machine learning models.

Key considerations in model evaluation and selection include:

1. **Data Splitting**: The dataset is typically split into two or more subsets: a training set used to train the model, a validation set used for hyperparameter tuning, and a test set used to evaluate the final model. Cross-validation techniques can also be employed for more robust evaluation.

2. **Performance Metrics**: Select appropriate performance metrics based on the type of problem. Common metrics include accuracy, precision, recall, F1-score for classification, and mean squared error (MSE), R-squared for regression.

3. **Hyperparameter Tuning**: Experiment with different hyperparameters (e.g., learning rates, regularization strengths) to optimize the model's performance. This is often done using techniques like grid search or random search.

4. **Cross-Validation**: Cross-validation is a technique for estimating the model's performance using different subsets of the data. Common methods include k-fold cross-validation and stratified sampling.

5. **Bias-Variance Trade-Off**: Balance the model's bias and variance to avoid underfitting (high bias) and overfitting (high variance). Choose a model complexity that generalizes well to unseen data.

6. **Ensemble Methods**: Explore ensemble methods such as bagging (e.g., random forests) and boosting (e.g., gradient boosting) to combine the strength of multiple models for improved performance.

7. **Feature Selection and Engineering**: Careful feature selection and engineering can improve model performance. It involves selecting the most relevant features and creating new features that might be more informative.

8. **Bias and Fairness**: Assessing and mitigating bias and ensuring fairness in ML models is essential, particularly in applications with societal impact.

9. **Interpretability and Explainability**: In some applications, model interpretability is critical. Techniques like LIME (Local Interpretable Model-agnostic Explanations) and SHAP (SHapley Additive exPlanations) can help explain model predictions.

10. **Regularization**: Techniques like L1 and L2 regularization can help prevent overfitting and improve model generalization.

11. **Validation Curves and Learning Curves**: Plot validation curves to understand how hyperparameters affect the model's performance. Learning curves help assess the model's performance as the dataset size increases.

12. **Comparison of Multiple Models**: Evaluating multiple models is important to choose the best-performing one. Visualization tools and model selection criteria like AIC (Akaike Information Criterion) and BIC (Bayesian Information Criterion) can help in this process.
 
13. **Model Selection**: After evaluating multiple models, choose the one that provides the best trade-off between accuracy and generalization.

14. **A/B Testing**: In some cases, deploying the model in a controlled A/B testing environment helps assess its real-world impact before full-scale deployment. Assess the final model's performance on the test set, ensuring that it can handle unseen data effectively.

15. **Deployment and Continuous Monitoring**: Prepare the model for deployment in a real-world application. After deployment, continuous monitoring and model retraining are essential to adapt to changing data distributions and maintain model performance.

Model evaluation and selection involve a mix of domain expertise, experimentation, and data-driven decision-making. The goal is to build models that perform well on unseen data and provide actionable insights or predictions.

## Accuracy with Imbalanced Classes
- Suppose you have two classes:
> - Relevant (R): the positive class
> - Not_Relevant (N): the negative class
- Out of 1000 randomly selected items, on average
> - One item is relevant and has an R label
> - The rest of the items (999 of them) are not relevant and labelled N.
- Recall that:<br>
Accuracy = (no. of correct predictions) / (no. of total instances)
- For comparison, suppose we had a "dummy" classifier that didn't look at the features at all, and always just blindly predicted the most frequent class (i.e. the negative N class).
- Assuming a test set of 1000 instances, what would this dummy classifier's accuracy be?
> - Answer : <br>
DUMMY Accuracy = 999 / 1000 = 99.9%
- Hence accuracy won't be correct metrics for measuring success of above case

### Dummy Classifiers

**DummyClassifier** is a classifier that makes predictions using simple rules, which can be useful as a baseline for comparison against actual classifiers, especially with imbalanced classes.  
Some commonly-used settings for the `strategy` parameter for DummyClassifierin scikit-learn:
- **most_frequent**: predicts the most frequent label in the training set.
- **stratified**: random predictions based on training set class distribution.
- **uniform**: generates predictions uniformly at random.
- **constant**: always predicts a constant label provided by the user.
> - A major motivation of this method is <u>F1-scoring</u>, when the positive class is in the minority.

```python
from sklearn.dummy import DummyClassifier

# Negative class (0) is most frequent
dummy_majority = DummyClassifier(strategy = 'most_frequent').fit(X_train, y_train)

# Therefore the dummy 'most_frequent' classifier always predicts class 0
y_dummy_predictions = dummy_majority.predict(X_test)

y_dummy_predictions
```

```python
dummy_majority.score(X_test, y_test)
```

```python
# Changing the kernel
svm = SVC(kernel='linear', C=1).fit(X_train, y_train)
svm.score(X_test, y_test)
```

## Null accuracy 
**Null accuracy** is a concept used in the evaluation of classification models in machine learning. It represents the accuracy that a model would achieve by always predicting the most frequent class in the dataset, effectively making it a "null" or "baseline" model. Calculating null accuracy is essential because it provides a benchmark for evaluating the performance of your model. Specifically, it helps you determine whether your model is making predictions better than random chance.

The null accuracy is calculated using the following formula:

Null Accuracy = (Number of samples in the most frequent class) / (Total number of samples)

Comparing the accuracy of a machine learning model to the null accuracy is a way to assess whether the model is learning meaningful patterns from the data. If the model's accuracy is significantly better than the null accuracy, it suggests that the model is making valuable predictions. However, if the model's accuracy is close to or worse than the null accuracy, it may indicate that the model is not performing well, and further improvements or model development are required.

### What if my classifier accuracy is close to the null accuracy baseline? 
This could be a sign of:
- Ineffective, erroneous or missing features
- Poor choice of kernel or hyperparameter
- Large class imbalance

## Confusion Matrices

A confusion matrix is a tool used in classification tasks to evaluate the performance of a machine learning model. It provides a clear and detailed summary of the model's predictions by comparing them to the actual ground truth values. Confusion matrices are especially useful when dealing with imbalanced datasets or when different errors have varying costs or consequences

In a confusion matrix, we have four essential components:

- **True Positives (TP)**: The number of instances correctly predicted as the positive class.

- **True Negatives (TN)**: The number of instances correctly predicted as the negative class.

- **False Positives (FP)**: The number of instances incorrectly predicted as the positive class (<u>a type I error</u>).

- **False Negatives (FN)**: The number of instances incorrectly predicted as the negative class (<u>a type II error</u>).


## Binary (two-class) confusion matrix

Compute confusion matrix to evaluate the accuracy of a classification.

By definition a confusion matrix `C` is such that `C(i,j)` is equal to the number of observations known to be in group `i` and predicted to be in group `j`. <br>
Thus in binary classification, the count of : <br>
- True Negatives - `C(0,0)`
- False Negatives - `C(1,0)`
- True Positives - `C(1,1)`
- False Positives - `C(0,1)`

### Confusion matrix for `DummyClassifier`
#### `strategy = 'most_frequent'`

```python
from sklearn.metrics import confusion_matrix

# Negative class (0) is most frequent
dummy_majority = DummyClassifier(strategy = 'most_frequent').fit(X_train, y_train)
y_majority_predicted = dummy_majority.predict(X_test)
confusion = confusion_matrix(y_test, y_majority_predicted)

print('Most frequent class (dummy classifier)\n', confusion)
```

#### `strategy = 'stratified'`

```python
# produces random predictions with same class proportion as training set
dummy_classprop = DummyClassifier(strategy='stratified').fit(X_train, y_train)
y_classprop_predicted = dummy_classprop.predict(X_test)
confusion = confusion_matrix(y_test, y_classprop_predicted)

print('Random class-proportional prediction (dummy classifier)\n', confusion)
```

### Confusion matrix for `SVC (kernel='linear')`

```python
svm = SVC(kernel='linear', C=1).fit(X_train, y_train)
svm_predicted = svm.predict(X_test)
confusion = confusion_matrix(y_test, svm_predicted)

print('Support vector machine classifier (linear kernel, C=1)\n', confusion)
```

### Confusion matrix for `LogisticRegression`

# Video 2 - Confusion Matrices & Basic Evaluation Metrics

Confusion matrices are valuable for calculating various classification metrics, including:

- **Accuracy**: The ratio of correctly predicted instances to the total number of instances.
> - (TP + TN) / (TP + TN + FP + FN)
- **Precision (Positive Predictive Value)**: The ratio of true positives to the total number of instances predicted as positive.
> - TP / (TP + FP)
- **Recall (Sensitivity or True Positive Rate [TPR] or Probablity of detection)**: The ratio of true positives to the total number of actual positive instances.
> - TP / (TP + FN)
- **F1 Score**:<br> It combines precision and recall into a single value. It's particularly useful when there is an uneven class distribution (class imbalance) in the dataset. <br>The F1 Score is the harmonic mean of precision and recall. It provides a balance between the two metrics. A high F1 Score indicates that the model has both good precision and recall, while a low F1 Score suggests that the model is lacking in either precision or recall.
> - 2 * (Precision * Recall) / (Precision + Recall)
- **Classification Error**: The ratio of incorrectly predicted instances to the total number of instances.
> - (1 - Accuracy)
> - (FP + FN) / (TP + TN + FP + FN)
- **False Positive Rate (FPR)**: The ratio of false positives to the total number of actual negative instances.
> - FP / (TN + FP)
- **Specificity (True Negative Rate [TNR])**: The ratio of true negatives to the total number of actual negative instances.
> - TN / (TN + FP)
> - (1 - FPR)

These metrics help in assessing the model's performance, particularly in binary classification tasks. Confusion matrices are also the basis for calculating receiver operating characteristic (ROC) curves and area under the curve (AUC) in binary and multiclass classification.

## How to decide what metric to apply ?

When deciding which evaluation metrics to apply in your machine learning project, it's essential to consider the nature of the problem, the dataset, and the specific goals. Here's a breakdown of when to use some common metrics:

1. **Accuracy**:
   - Use when you want a simple and intuitive measure of <u>overall model performance</u>.
   - Suitable for <u>balanced datasets</u>, where classes are relatively evenly distributed.
   - <u>Not</u> recommended for highly imbalanced datasets, as it can be misleading.


2. **Precision**:
   - Use when <u>minimizing false positives</u> is a priority.
   - Important in situations where false positives are costly or detrimental.
   - Examples include medical diagnoses and fraud detection.
   - <u>Scenario</u> - Imagine police use an algo for finding likely criminals in a given area. In this case, a FP (arresting innocent person) may be considered more damaging than a FN (letting a potential criminal walk free). Hence this is <u>precision-oriented</u> task that should minimize FPs


3. **Recall**:
   - Use when <u>minimizing false negatives</u> is critical.
   - Important when missing positive cases (false negatives) has severe consequences.
   - Useful in applications like disease detection or security.
   - <u>Scenario</u> - A law firm is trying to find all emails that mention a certain event (discovery process). Missing even one IMP email (FN) could omit valuable evidence. It's OK to include FPs (emails that are actually not relevant) because we have experts who can filter them out later. Hence this is <u>recall-oriented </u> task that should minimize FNs


4. **F1 Score**:
   - Use when you need a <u>balance</u> between precision and recall.
   - Helpful when there's a trade-off between false positives and false negatives.
   - Especially valuable in <u>imbalanced</u> datasets.


5. **Classification Error**:
   - Use when you want a straightforward measure of classification errors.
   - Not as commonly used as accuracy but can be informative.
   - More suitable for smaller datasets.


6. **Specificity**:
   - Use when distinguishing true negatives is important.
   - Particularly relevant in applications where you want to <u>minimize false alarms</u>.
   - Complements sensitivity (recall).


7. **False Positive Rate (FPR)**:
   - Use when understanding the rate of false alarms is essential.
   - Useful for applications where you want to control and <u>minimize the false positive</u> rate.

The choice of metric depends on your problem context, class distribution, and the relative costs of false positives and false negatives. It's common to use multiple metrics to gain a holistic view of model performance. Additionally, you can visualize metrics like precision-recall curves and ROC curves to understand trade-offs better and make informed decisions.

## Tradeoff between precision and recall

In machine learning, there is often a tradeoff between precision and recall when evaluating classification models. These two metrics are inversely related and understanding this tradeoff is crucial in choosing the right balance for your specific application.

1. **Precision**:
   - Precision measures the accuracy of positive predictions made by the model. It answers the question, "Of the instances predicted as positive, how many are truly positive?"
   - High precision means that when the model predicts a positive class, it's likely to be correct.
   - Precision is particularly important in situations where false positives are costly or detrimental, such as in spam email filtering.

2. **Recall**:
   - Recall (also known as Sensitivity) measures the model's ability to identify all relevant instances of the positive class. It answers the question, "Of all the true positive instances, how many did the model correctly identify?"
   - High recall means that the model is good at capturing most of the positive instances in the dataset.
   - Recall is crucial when missing positive cases (false negatives) has severe consequences, such as in disease detection or security.

The tradeoff between precision and recall occurs because, in practice, <u>increasing one metric often results in the decrease of the other</u>. Here's how this tradeoff works:

- **Increasing Precision**: You can achieve higher precision by becoming more conservative with positive predictions. This means making fewer positive predictions, but when you do, they are more likely to be correct. However, this approach can result in a lower recall because some true positive instances might be missed.

- **Increasing Recall**: You can increase recall by becoming more liberal with positive predictions, capturing more potential positive instances. This might lead to a higher number of true positives, but it can also increase the number of false positives, resulting in lower precision.

The choice between precision and recall depends on your specific goals and the nature of the problem:

- If you want to <u>minimize false positives</u> (avoid false alarms), you should prioritize <u>precision</u>.
- If you want to capture as <u>many true positive cases</u> as possible, even at the cost of <u>some false alarms</u>, you should prioritize <u>recall</u>.

The ideal balance between precision and recall will vary from one application to another. For ex. - <br>
**Precision-oriented ML tasks**:
- Search engine ranking, query suggestion
- Document classification
- Many customer-facing tasks (users remember failures!)
  

**Recall-oriented ML tasks:**
- Search and information extraction in legal discovery
- Tumor detection
- Often paired with a human expert to filter out false positives

In practice, it's common to consider <u>both metrics simultaneously</u> and use a measure that combines them, such as the <u>F1 Score</u>, to find a suitable tradeoff that aligns with the objectives of the task.

## F-Score vs. F1 Score

The terms "F-Score" and "F1 Score" are often used interchangeably, but there is a subtle difference in their definitions. Both metrics are designed to provide a balance between precision and recall in binary classification tasks. Let's clarify their meanings:

1. **F-Score**:
   - The term "F-Score" is a more general concept that represents a family of metrics that balance precision and recall. It is expressed as Fβ, where β is a parameter that determines the weight given to precision or recall.
   - The general formula for Fβ is: Fβ = (1 + β^2) * (Precision * Recall) / (β^2 * Precision + Recall).
   - <u>When β = 1</u>, you get the F1 Score, which is the harmonic mean of precision and recall.
   - 

2. **F1 Score**:
   - The F1 Score is a specific case of the F-Score where β is set to 1 and hence 1 in F1 means β=1. It is the harmonic mean of precision and recall, calculated as <br> F1 Score = 2 * (Precision * Recall) / (Precision + Recall).
   - The F1 Score is the most commonly used F-Score when you want an equal balance between precision and recall.

**Key Differences**:

- The F1 Score, with β = 1, is a special case of the F-Score and is widely used in machine learning.
- F1 Score has equal weight for precision and recall, making it suitable when you want a balance between minimizing false positives and false negatives.
- F-Score with different values of β allows you to adjust the balance between precision and recall based on the specific requirements of your task. <br> For example, <u>F2 Score gives more weight to recall, while F0.5 Score favors precision.</u>

**Use Cases**:

- F1 Score is commonly used in many binary classification scenarios, especially when you want to <u>balance</u> the trade-off between precision and recall.
- Specific F-Scores (e.g., F2 Score or F0.5 Score) can be used when there are different priorities for precision and recall in a specific application.

In summary, both F-Score and F1 Score are valuable metrics for balancing precision and recall in classification tasks. The F1 Score is the most widely used and represents an equal balance, while F-Scores with different β values allow customization of the balance between these two metrics.

## `decision_function()` Method in scikit-learn

The `decision_function()` method is a function available in scikit-learn's machine learning library. It is primarily used with various classifiers, such as Support Vector Machines (SVM), to provide insight into how the classifier makes decisions and assigns class labels. The method calculates and returns decision scores or distances for each instance in the dataset, which can be helpful for understanding the classifier's behavior.

**Key Points**:

1. **Output Type**:
   - The `decision_function()` method typically returns a one-dimensional array of decision scores or distances for each instance in the dataset. The interpretation of these scores may vary depending on the classifier used.
   - Each classifier score value per test point indicates how confidently the classifier predicts the positive class (large-magnitude positive values) or the negative class (large-magnitude negative values).
   - Choosing a fixed decision threshold gives a classification rule.
   - By sweeping the decision threshold through the entire range of possible score values, we get a series of classification outcomes that form a curve.


2. **Use Cases**:
   - The method is often used to assess the confidence or strength of the classifier's decisions. For example, in a binary classification scenario, positive decision scores might indicate high confidence in the positive class, while negative scores suggest confidence in the negative class.


3. **Thresholding**:
   - Decision scores are continuous values. To make binary predictions, a threshold is applied to these scores. Instances with scores above the threshold are assigned to one class, and those below the threshold are assigned to the other class.


4. **Classifier-specific**:
   - The interpretation of the decision scores can vary from one classifier to another. For example:
     - In Support Vector Machines (SVM), the decision scores represent the signed distances of instances to the separating hyperplane. The sign (+ or -) determines the assigned class.
     - In Logistic Regression, the decision scores correspond to the estimated probabilities of belonging to the positive class.

**Useful Applications**:

- Model Evaluation: The `decision_function()` method can be used to evaluate the performance of a classifier, especially when precision-recall trade-offs or ROC curves are analyzed. Decision scores help understand the trade-off between precision and recall by adjusting the decision threshold.

- Anomaly Detection: In some applications, decision scores can be used for anomaly or outlier detection. Instances with low scores might be flagged as anomalies.

- Model Interpretation: Decision scores can provide insights into why a model made a particular prediction. For example, SVM decision scores can reveal how close an instance is to the decision boundary.

## Precision-Recall Curves

Precision-recall curves are a valuable tool for evaluating and visualizing the performance of classification models, particularly in scenarios with imbalanced datasets or when there's a need to prioritize specific aspects of model performance, such as precision or recall.

**Key Components**:

1. **Precision**:
   - Precision is a measure of how accurate the positive predictions of a model are. It quantifies the proportion of true positive predictions among all positive predictions.
   - Precision = True Positives / (True Positives + False Positives)


2. **Recall** (Sensitivity):
   - Recall, also known as sensitivity or true positive rate, measures the model's ability to capture all relevant instances of the positive class. It calculates the proportion of true positives among all actual positive instances.
   - Recall = True Positives / (True Positives + False Negatives)

**Precision-Recall Curve**:

- A precision-recall curve is created by varying the decision threshold of a classification model and calculating precision and recall at each threshold. These values are plotted on a graph, where different points on the curve correspond to different thresholds.

- As the threshold changes, precision and recall values also change. When the threshold is low, the model is more likely to make positive predictions, resulting in higher recall but lower precision. Conversely, when the threshold is high, precision increases but recall decreases.

**Use Cases**:

1. **Imbalanced Datasets**: Precision-recall curves are particularly useful when dealing with imbalanced datasets, where one class is significantly larger than the other. In such cases, accuracy can be misleading, and precision-recall analysis provides a more informative view of model performance.

2. **Threshold Selection**: Precision-recall curves help in selecting an optimal threshold for a specific task. The choice of threshold depends on the relative importance of precision and recall for that application.

3. **Model Comparison**: Precision-recall curves enable the comparison of multiple classifiers. You can visually compare the trade-offs between precision and recall for different models and select the one that best aligns with your objectives.

4. **Anomaly Detection**: In anomaly detection, precision-recall curves are useful for selecting a threshold that minimizes false positives while maximizing the detection of rare anomalies.

**Interpretation**:

- A model's performance is better if its precision-recall curve is closer to the upper-right corner of the graph, as this signifies higher precision and recall values for a broader range of thresholds.
    - Top right corner:
    - The “ideal” point
    - Precision = 1.0
    - Recall = 1.0

- The area under the precision-recall curve (AUC-PR) can also be used as a single value to quantify the overall performance of the model. A higher AUC-PR indicates better model performance.

- “Steepness” of P-R curves is important:
    - Maximize precision
    - while maximizing recall

In summary, precision-recall curves are a powerful tool for assessing the performance of classification models, especially in situations where class imbalance and the trade-offs between precision and recall need to be considered. They provide valuable insights for selecting thresholds, model comparison, and optimizing model behavior to align with specific application requirements.

## ROC curves, Area-Under-Curve (AUC)

**Receiver Operating Characteristic (ROC)** curves and the Area Under the Curve (AUC) are important tools for evaluating and visualizing the performance of binary classification models. They provide valuable insights into a model's ability to discriminate between positive and negative instances.

**Key Components**:

1. **True Positive Rate (TPR) (Recall)**:
   - TPR, also known as recall or sensitivity, measures the proportion of true positive predictions among all actual positive instances.
   - TPR = True Positives / (True Positives + False Negatives)
   

2. **False Positive Rate (FPR)**:
   - FPR quantifies the proportion of false positive predictions among all actual negative instances.
   - FPR = False Positives / (False Positives + True Negatives)

**ROC Curve**:

- An ROC curve is a graphical representation of the relationship between TPR and FPR as the decision threshold of a classification model is varied. It is created by plotting TPR (y-axis) against FPR (x-axis) at different threshold values.

- The ROC curve illustrates how well a model can distinguish between positive and negative instances across various threshold settings. A curve closer to the upper-left corner indicates better performance.

- The “ideal” point :
    - Top left corner
    - False positive rate of zero
    - True positive rate of one

- “Steepness” of ROC curves is important:
    - Maximize the true positive rate
    - while minimizing the false positive rate

```python
from sklearn.metrics import roc_curve, auc

X_train, X_test, y_train, y_test = train_test_split(X, y_binary_imbalanced, random_state=0)

y_score_lr = lr.fit(X_train, y_train).decision_function(X_test)
fpr_lr, tpr_lr, _ = roc_curve(y_test, y_score_lr)
roc_auc_lr = auc(fpr_lr, tpr_lr)

plt.figure()
plt.xlim([-0.01, 1.00])
plt.ylim([-0.01, 1.01])
plt.plot(fpr_lr, tpr_lr, lw=3, label='LogRegr ROC curve (area = {:0.2f})'.format(roc_auc_lr))
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate\n(Recall)', fontsize=12)
plt.title('ROC curve (1-of-10 digits classifier)', fontsize=16)
plt.plot([0, 1], [0, 1], label = 'Baseline (Random Classifier ROC Curve)', color='navy', lw=3, linestyle='--')
plt.legend(loc='lower right', fontsize=11)
# plt.axes().set_aspect('equal')
plt.show()
```

- **Macro-averaging:** calculates the metrics independently for each class and then averages them. It treats all classes equally.


```python
print('Micro-averaged precision = {:.2f} (treat instances equally)'
      .format(precision_score(y_test_mc, svm_predicted_mc, average = 'micro')))
print('Macro-averaged precision = {:.2f} (treat classes equally)'
      .format(precision_score(y_test_mc, svm_predicted_mc, average = 'macro')))
```

```python
print('Micro-averaged f1 = {:.2f} (treat instances equally)'
      .format(f1_score(y_test_mc, svm_predicted_mc, average = 'micro')))
print('Macro-averaged f1 = {:.2f} (treat classes equally)'
      .format(f1_score(y_test_mc, svm_predicted_mc, average = 'macro')))
```

## Micro-Average vs Macro-Average
- If the classes have about the same number of instances, micro- and macro-average will be about the same.
- If some classes are much larger (more instances) than others, and you want to:
    - Weight your metric toward the largest ones, use micro-averaging.
    - Weight your metric toward the smallest ones, use macro-averaging.
- If the micro-average is much lower than the macro-average then examine the larger classes for poor metric performance.
- If the macro-average is much lower than the micro-average then examine the smaller classes for poor metric performance.

## Multi-Label Classification Evaluation

Multi-label classification is a classification problem where each instance can be assigned to multiple classes or categories. When evaluating the performance of multi-label classification models, specific metrics and techniques are employed to measure their effectiveness.

**Key Metrics**:

1. **Hamming Loss**:
   - Hamming Loss is a metric that measures the fraction of labels that are incorrectly predicted across all instances. It quantifies the overall accuracy of multi-label classification.


2. **Precision@K and Recall@K**:
   - Precision@K and Recall@K are metrics that consider the top-K predicted labels for each instance. They assess how well the model ranks the correct labels within the top-K predictions.


3. **F1-Score**:
   - F1-Score is the harmonic mean of precision and recall for multi-label classification. It provides a balance between precision (the accuracy of positive predictions) and recall (the model's ability to capture all relevant labels).


4. **Coverage**:
   - Coverage measures the average number of labels needed to include in the top-K predictions to cover all the true labels. A lower coverage value indicates better performance.


5. **One Error**:
   - One Error evaluates the percentage of instances for which the top predicted label is incorrect. A lower one error is desirable.

**Use Cases**:
- Multi-label classification is commonly applied to various tasks, such as image tagging, document categorization, music genre classification, and recommendation systems.
- It helps solve problems where instances can belong to multiple categories simultaneously.


**Challenges**:
- Multi-label classification evaluation is more complex than traditional single-label classification due to the presence of multiple labels per instance.
- The choice of evaluation metrics depends on the application and the importance of precision, recall, or other factors.


**Interpretation**:
- Hamming Loss provides an overall assessment of how well the model predicts labels.
- Precision@K and Recall@K consider the top-K predictions for each instance and assess how well the model ranks the correct labels.
- F1-Score balances precision and recall, making it valuable for multi-label evaluation.
- Coverage and One Error provide insights into the coverage and ranking accuracy of the top-K predictions.

In summary, multi-label classification evaluation involves specific metrics and techniques tailored to problems where instances can have multiple associated labels. These metrics assess the model's ability to predict the correct labels, rank them accurately, and balance precision and recall for multi-label classification tasks.

# Video 6 - Regression evaluation metrics

Regression is a type of supervised learning where the goal is to predict a continuous numerical value, making it distinct from classification. Several evaluation metrics are used to assess the performance of regression models, measuring the accuracy and goodness of fit.

**Key Metrics**:

1. **Mean Absolute Error (MAE)**:
   - MAE quantifies the mean absolute difference between the actual values and the predicted values. It provides a straightforward measure of model accuracy.
   - MAE = (1/n) * Σ|actual - predicted|
   

2. **Mean Squared Error (MSE)**:
   - MSE calculates the mean squared difference between the actual values and the predicted values. It emphasizes the impact of larger errors.
   - MSE = (1/n) * Σ(actual - predicted)^2
   

3. **Root Mean Squared Error (RMSE)**:
   - RMSE is the square root of MSE, providing a metric with the same unit of measurement as the target variable. It is widely used in practice.
   - RMSE = √MSE =  √[(1/n) * Σ(actual - predicted)^2]
   

4. **Mean Absolute Percentage Error (MAPE)**:
   - MAPE calculates the mean percentage difference between predicted and actual values, making it useful for understanding relative error.
   - MAPE = (1/n) * Σ(|(actual - predicted) / actual| * 100)
   

5. **R-squared (R²) Score**:
   - R-squared measures the proportion of the variance in the dependent variable that can be explained by the independent variables in the model. It quantifies the goodness of fit.
   - R² = 1 - (SSE / SST), where SSE is the sum of squared residual error, and SST is the total sum of squares.


6. **Adjusted R-squared**:
   - Adjusted R-squared is an extension of R-squared that takes into account the number of predictors in the model, adjusting for model complexity.
   
   
7. **Median Absolute Error (MedAE)**:
    - Median Absolute Error (MedAE) is a robust metric used for evaluating the performance of regression models. Unlike mean-based metrics such as Mean Absolute Error (MAE) or Mean Squared Error (MSE), MedAE focuses on the median of the absolute errors, making it less sensitive to outliers and extreme values in the dataset.
    - MedAE = Median(|actual - predicted|)
    - **Robustness**: MedAE is less influenced by outliers, anomalies, or extreme values in the dataset. This makes it a valuable metric in situations where the dataset contains data points that significantly deviate from the majority. It does not overly penalize models for occasional extreme prediction errors. 
    - While MedAE is robust to outliers, it may not be as sensitive to smaller, more common errors.
    - MedAE is particularly useful in regression tasks where the dataset may have outliers or when the model's performance on typical instances is more critical than its performance on extreme cases.
    - A lower MedAE indicates that, on average, the model's predictions are closer to the actual values, with less deviation.
    

**Interpretation**:

- Lower values of MAE, MSE, RMSE and MedAE indicate better model accuracy, with lower errors.
- MAPE expresses errors as percentages relative to actual values, providing a sense of the magnitude of errors.
- R-squared measures how well the model explains the variation in the target variable. An R² value close to 1 suggests a good fit.

**Typically r2_score is enough for regression**
- <u>Reminder:</u> computes how well future instances will be predicted
- Best possible score is 1.0
- Constant prediction score is 0.0


## R-squared (R²) Score

The **R-squared (R²)** score, also known as the <u>coefficient of determination</u>, is a common metric used in <u>regression analysis</u> to assess the goodness of fit of a regression model. It quantifies the proportion of the variance in the dependent variable that can be explained by the independent variables in the model. The R² score provides insights into how well the model captures the variability in the data and how well it predicts the target variable.

**Formula**:

The R² score is calculated using the following formula:

R² = 1 - (SSE / SST) = SSR / SST           ....because SST = SSR + SSE

Where:
- SSE (Sum of Squared Residual Error) is the sum of the squared differences between the actual values and the predicted values from the regression model.
- SST (Total Sum of Squares) is the sum of the squared differences between the actual values and the mean of the actual values.
- SSR (Sum of Squared Regression) is the sum of squared differences between the predicted values generated by the regression model and the mean of the the actual values. 

The R² score range:

- An R² score of 0 indicates that the model does not explain any of the variance in the dependent variable, and its predictions are no better than using the <u>mean</u> of the dependent variable.
- An R² score of 1 indicates that the model perfectly explains all the variance, and its predictions align with the actual data points.
- R² score can be negative

**Interpretation**:

- An R² score close to 0 indicates that the model does not capture the variance in the data, implying a poor fit.
- An R² score close to 1 suggests that the model's predictions closely match the actual data points, indicating a good fit.

**Use Cases**:

- R² is widely used in regression analysis to evaluate the quality of models, including linear regression, polynomial regression, and other regression techniques.
- It helps determine whether the independent variables in the model have a significant impact on the dependent variable.

**Limitations**:

- R² does not provide information about the statistical significance of the coefficients or the quality of individual predictors.
- A high R² score does not necessarily mean the model is accurate or that it generalizes well to new data. It may be overfitting the training data.

In summary, the R-squared score is a valuable metric for assessing the goodness of fit in regression models. It quantifies the proportion of variance in the dependent variable explained by the model's independent variables, providing insights into its predictive power and the quality of the fit.

```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.dummy import DummyRegressor

diabetes = datasets.load_diabetes()

X = diabetes.data[:, None, 6]
y = diabetes.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

lm = LinearRegression().fit(X_train, y_train)
lm_dummy_mean = DummyRegressor(strategy = 'mean').fit(X_train, y_train)

y_predict = lm.predict(X_test)
y_predict_dummy_mean = lm_dummy_mean.predict(X_test)

print('Linear model, coefficients: ', lm.coef_)
print("Mean squared error (dummy): {:.2f}".format(mean_squared_error(y_test, 
                                                                     y_predict_dummy_mean)))
print("Mean squared error (linear model): {:.2f}".format(mean_squared_error(y_test, y_predict)))
print("r2_score (dummy): {:.2f}".format(r2_score(y_test, y_predict_dummy_mean)))
print("r2_score (linear model): {:.2f}".format(r2_score(y_test, y_predict)))

# Plot outputs
plt.scatter(X_test, y_test,  color='black')
plt.plot(X_test, y_predict, color='green', linewidth=2)
plt.plot(X_test, y_predict_dummy_mean, color='red', linestyle = 'dashed', 
         linewidth=2, label = 'dummy')

plt.show()
```

## SSR SST SSE
**SSR (Sum of Squared Regression), SST (Total Sum of Squares), and SSE (Sum of Squared Errors) in Regression Evaluation**

In regression analysis, SSR (Sum of Squared Regression), SST (Total Sum of Squares), and SSE (Sum of Squared Errors) are important components of various evaluation metrics, helping to understand the distribution of variance and the model's ability to explain observed data.

**SSR (Sum of Squared Regression)**:

- SSR represents the sum of squared differences between the predicted values generated by the regression model and the mean of the dependent variable. It quantifies the portion of the total variation in the dependent variable that is explained by the model.

- Mathematically, SSR is calculated as follows:
  - SSR = Σ(predicted - mean(actual))² for all data points

**SST (Total Sum of Squares)**:

- SST, also known as the total sum of squares, represents the total variance in the dependent variable (target) without considering the influence of the regression model. It quantifies the spread of the actual target values around their mean.

- Mathematically, SST is calculated as follows:
  - SST = Σ(actual - mean(actual))² for all data points

**SSE (Sum of Squared Errors)**:

- SSE represents the sum of squared differences between the actual target values and the predicted values by the regression model. It quantifies the unexplained variation or the residual variance.

- Mathematically, SSE is calculated as follows:
  - SSE = Σ(actual - predicted)² for all data points
  


**R-squared (R²) Score**:

- The R-squared (R²) score, a commonly used metric in regression evaluation, is calculated as the ratio of SSR to SST:
  - R² = 1 - (SSR / SST)

**Interpretation**:

- SSR measures the variance in the target variable that is explained by the regression model. It represents the portion of the variance that the model can account for.

- SST represents the total variance in the target variable, irrespective of the model. It reflects the variability in the actual target values.

- SSE measures the unexplained variance, or the portion of variance not captured by the model. It represents the difference between the actual values and the model's predictions.

- R², calculated from SSR, SST, and SSE, quantifies the proportion of the variance in the dependent variable that is explained by the independent variables in the regression model.

- Mathematically, **SST = SSR + SSE**.

**Use Cases**:

- SSR, SST, and SSE, along with the R² score, are used to assess the quality of a regression model and its ability to explain the variation in the dependent variable.

- They provide insights into how well the model captures the underlying relationships between the independent and dependent variables.

In summary, SSR (Sum of Squared Regression), SST (Total Sum of Squares), and SSE (Sum of Squared Errors) are important components in regression evaluation, helping to understand the distribution of variance and the model's ability to explain the observed data. The R-squared score, calculated from these components, provides a key metric for assessing the goodness of fit in regression models.

## Grid search example
**GridSearchCV in scikit-learn: Hyperparameter Tuning Made Easy**

Hyperparameter tuning is a crucial part of building machine learning models. It involves finding the best combination of hyperparameters for a given algorithm to optimize its performance. GridSearchCV in scikit-learn is a powerful tool that simplifies this process.

**What is GridSearchCV?**

GridSearchCV, short for Grid Search Cross-Validation, is a hyperparameter optimization technique in scikit-learn. It systematically explores a predefined hyperparameter grid to find the best combination that yields the highest model performance.

**Key Components of GridSearchCV**:

1. **Hyperparameter Grid**: You define a grid of hyperparameters that you want to search through. This grid consists of possible values or ranges for hyperparameters.

2. **Cross-Validation**: GridSearchCV performs cross-validation on different combinations of hyperparameters. Cross-validation helps ensure that the model's performance is evaluated on multiple subsets of the training data, reducing the risk of overfitting.

3. **Scoring Function**: You specify the scoring function you want to optimize. Common choices include accuracy, F1-score, or other evaluation metrics.

**How GridSearchCV Works**:

1. GridSearchCV takes the hyperparameter grid, cross-validation configuration, and scoring function as input.

2. It systematically evaluates the model's performance for each combination of hyperparameters by fitting the model to the training data and evaluating it on the validation data for each fold of the cross-validation.

3. The combination of hyperparameters that results in the highest score (based on the chosen scoring function) is selected as the best.

4. Once the best hyperparameters are found, GridSearchCV can be used to train the model with those hyperparameters on the full training dataset.

**Benefits of GridSearchCV**:

- **Automation**: GridSearchCV automates the hyperparameter tuning process, saving time and effort. You don't need to manually test various hyperparameter combinations.

- **Optimization**: It helps in finding the best set of hyperparameters for your specific problem and dataset.

- **Reduced Overfitting**: By using cross-validation, it reduces the risk of overfitting because the model is evaluated on multiple validation sets.

- **Consistency**: GridSearchCV ensures that your experiments are consistent and reproducible, as it systematically explores the hyperparameter space.

**Considerations**:

- GridSearchCV can be computationally expensive, especially with a large hyperparameter grid. Be mindful of the time and resources it may require.

- It's crucial to choose the right hyperparameter grid and scoring function. Poor choices can lead to suboptimal results.

- GridSearchCV does not guarantee the best hyperparameters for all problems. Some problems may benefit from more advanced techniques like random search or Bayesian optimization.

In summary, GridSearchCV in scikit-learn is a valuable tool for hyperparameter tuning, making it easier to find the best hyperparameters for your machine learning models. It automates the search process, reduces overfitting risk, and ensures consistency in your experiments.


```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

dataset = load_digits()
X, y = dataset.data, dataset.target == 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

clf = SVC(kernel='rbf')
grid_values = {'gamma': [0.001, 0.01, 0.05, 0.1, 1, 10, 100]}

# default metric to optimize over grid parameters: accuracy
grid_clf_acc = GridSearchCV(clf, param_grid = grid_values)
grid_clf_acc.fit(X_train, y_train)
y_decision_fn_scores_acc = grid_clf_acc.decision_function(X_test) 

print('Grid best parameter (max. accuracy): ', grid_clf_acc.best_params_)
print('Grid best score (accuracy): ', grid_clf_acc.best_score_)

# alternative metric to optimize over grid parameters: AUC
grid_clf_auc = GridSearchCV(clf, param_grid = grid_values, scoring = 'roc_auc')
grid_clf_auc.fit(X_train, y_train)
y_decision_fn_scores_auc = grid_clf_auc.decision_function(X_test) 

print('Test set AUC: ', roc_auc_score(y_test, y_decision_fn_scores_auc))
print('Grid best parameter (max. AUC): ', grid_clf_auc.best_params_)
print('Grid best score (AUC): ', grid_clf_auc.best_score_)
```

### Precision-recall curve for the default SVC classifier (with 'balanced' `class weights`)

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve
from adspy_shared_utilities import plot_class_regions_for_classifier
from sklearn.svm import SVC

dataset = load_digits()
X, y = dataset.data, dataset.target == 1
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

# create a two-feature input vector matching the example plot above
jitter_delta = 0.25
X_twovar_train = X_train[:,[20,59]]+ np.random.rand(X_train.shape[0], 2) - jitter_delta
X_twovar_test  = X_test[:,[20,59]] + np.random.rand(X_test.shape[0], 2) - jitter_delta

clf = SVC(kernel='linear', class_weight='balanced').fit(X_twovar_train, y_train)

y_scores = clf.decision_function(X_twovar_test)

precision, recall, thresholds = precision_recall_curve(y_test, y_scores)
closest_zero = np.argmin(np.abs(thresholds))
closest_zero_p = precision[closest_zero]
closest_zero_r = recall[closest_zero]

plot_class_regions_for_classifier(clf, X_twovar_test, y_test)
plt.title("SVC, class_weight = 'balanced', optimized for accuracy")
plt.show()

plt.figure()
plt.xlim([0.0, 1.01])
plt.ylim([0.0, 1.01])
plt.title ("Precision-recall curve: SVC, class_weight = 'balanced'")
plt.plot(precision, recall, label = 'Precision-Recall Curve')

# the red circle represents the precision-recall tradeoff achieved at the zero score mark, 
# which is the actual decision boundary chosen for the trained classifier.
plt.plot(closest_zero_p, closest_zero_r, 'o', markersize=12, fillstyle='none', c='r', mew=3)

plt.xlabel('Precision', fontsize=16)
plt.ylabel('Recall', fontsize=16)
# plt.axes().set_aspect('equal')
plt.show()
print('At zero threshold, precision: {:.2f}, recall: {:.2f}'
      .format(closest_zero_p, closest_zero_r))
```

## Training, Validation, and Test Framework for Model Selection and Evaluation
- Using only cross-validation or a test set to do model selection may lead to more subtle overfitting / optimistic generalization estimates.
- Instead, use three data splits:
    1. Training set (model building)
    2. Validation set (model selection)
    3. Test set (final evaluation)
- In practice:
    - Create an initial training/test split
    - Do cross-validation on the training data for model/parameter selection
    - Save the held-out test set for final model evaluation
    
## Concluding Notes
- Accuracy is often not the right evaluation metric for many realworld machine learning tasks
    - False positives and false negatives may need to be treated very differently
    - Make sure you understand the needs of your application and choose an evaluation metric that matches your application, user, or business goals.
- Examples of additional evaluation methods include:
    - Learning curve: How much does accuracy (or other metric) change as a function of the amount of training data?
